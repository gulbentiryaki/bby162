```
from google.colab import drive
drive.mount('/gdrive')
dosya = "/gdrive/My Drive/Colab Notebooks/BBY162-2020/veritabani.txt"
def eser_ekle():
  dosya1 = open(dosya, "r+",encoding="utf-8")
  liste3=dosya1.readlines()
  num=len(liste3)+1
  kitap_adi = input("Kitap Adı: ")
  yazar_adi = input("Yazar Adı: ")
  basim_yili = input("Basım Yılı: ")
  isbn = input("isbn : ")
  yayin_evi = input("Yayın Evi: ")
  dosya1.write(str(num)+"|"+kitap_adi+"|" + yazar_adi + "|" + basim_yili + "|" + isbn + "|" + yayin_evi+"\n")
  print("Eser Kayıt edildi.")
  dosya1.close()

def eser_listele():
  dosya2 = open(dosya, "r", encoding='utf-8')
  print(dosya2.read())
  dosya2.close()
def eser_guncelle():
  dosya3 = open(dosya, "r+", encoding='utf-8')
  liste=dosya3.readlines()
  print(liste)
  eser_num_guncelle=int(input("Güncellemek İstediğniz Eserin Numarasını Giriniz:"))
  kitap_adi=input("Kitap Adı: ")
  yazar_adi = input("Yazar Adı: ")
  basim_yili = input("Basım Yılı: ")
  isbn = input("isbn : ")
  yayin_evi = input("Yayın Evi: ")
  liste[int(eser_num_guncelle)-1]=(str(eser_num_guncelle))+"|"+kitap_adi+"|" + yazar_adi + "|" + basim_yili + "|" + isbn + "|" + yayin_evi+"\n"
  dosya3.seek(0)
  dosya3.writelines(liste)
  print(liste)
  print("Güncelleme Tamamlandı!")


def eser_arama():
  dosya5=open(dosya, "r", encoding='utf-8')
  liste2=dosya5.readlines()
  eser_ara=int(input("Aramak İstediğiniz Eserin Numarasını Giriniz:"))
  print(liste2[eser_ara-1])



islem = 0
while  islem != 3 :
  print("Yapılacak işlemi Seçiniz")
  print("1-Eser ekle")
  print("2-Eserleri Listele")
  print("3-Eser Güncelleme")
  print("4-Eser Arama")
  print("5-Çıkış")
  
  islem = int (input("Seçilen İşlem Tipi:"))
  if islem == 1:
    eser_ekle()
  elif islem == 2:
    eser_listele()
  elif islem == 3:
    eser_guncelle()
  elif islem == 4:
    eser_arama()
  elif islem == 5:
    break
  
  else:
    print("Tekrar Seçiniz.")
    continue
```
